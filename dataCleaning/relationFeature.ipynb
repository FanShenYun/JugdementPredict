```python
import pandas as pd
import numpy as np
import re
import os, sys
pd.set_option('display.max_rows', 2500)
```

```python
df = pd.read_csv('./onlyBillText.csv',header=None,sep=',')
```

```python
df
```

```python
regex = r'[\u4E00-\u9FFF]{2}糾紛|[\u4E00-\u9FFF]{2}爭執|停車細故|[\u4E00-\u9FFF]{2}爭議|[\u4E00-\u9FFF]{2}糾葛|積欠[\u4E00-\u9FFF]{2}'
regex2 = r'鄰居|管理委員會|[\u4E00-\u9FFF]{2}關係|不認識|先遭[\u4E00-\u9FFF]+?提告|[男|女]朋友|夫妻|網路遊戲'
relateList=[]
j=0
for i in df[1]:
    zhText=''
    if type(i) == float:
        zhText=''
    elif re.findall(regex, i) != []:
        zhText = re.findall(regex, i)[0]
    elif re.findall(regex2, i) != []:
        zhText = re.findall(regex2, i)[0]
    relateList.append(zhText)
    j=j+1
        
relateList
```

```python
df[3]=relateList
```

```python
df2 = df.drop([1], axis=1)
```
