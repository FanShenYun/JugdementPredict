# 移除空白跟格式錯誤出現的欄位

```python
import pandas as pd
import numpy as np
import re
import os, sys
```

```python
# 讀原檔在的資料夾
dirs = os.listdir('.\JudgeText')

# 要用到兩個正規化，第一個是移除空白用的正規化，第二個是判斷這個欄位是不是可以移除的數字欄位以及擷取csv檔名
regex = r'[0-9A-Za-z\u4E00-\u9FFF，。：「」；、？！『』〈〉（）《》○【】]+'
regex2 = r'[0-9]+'

# 讀每個csv檔
for file in dirs:
    strText=""
    title = re.match(regex2,file)[0] # 擷取檔名
    df = pd.read_csv(f'./JudgeText/{file}',header=None,sep=',') # 注意header一定要設None
    for i in range(len(df.columns)): # 13~15行讀csv中每個儲存格，觀察上儲存格會有三種內容：判決文字、NaN、數字欄位
        text=df[i].tolist()
        for j in range(len(text)):
            # 用if判斷處理三種儲存格狀況
            if type(text[j]) == float:  # 如果是NaN，type會是float，NaN不寫入所以用continue跳過
                continue
            elif re.match(regex2,text[j]) == None: # 如果是判決文字，用regex2正規化會出現None的結果，所以出現None代表應該寫入
                strText += str(text[j])
            else: # 不是NaN，也不是判決文字，就是數字欄位，一樣是continue跳過
                continue

    # 抓取所有文字部分並合成為一個string
    zhText = re.findall(regex, strText)
    cleanText = ''.join(zhText)
    # 寫成txt，用UTF-8編碼
    path = f'./totalText/{title}.txt'
    f = open(path, 'w', encoding='UTF-8')
    print(cleanText, file=f)
    f.close()
```

```python
# 有些csv經過前面regex2會抓不到，需要人工抓取
cleanText=""
textAttend = []
strText=""
regex = r'[0-9A-Za-z\u4E00-\u9FFF，。：「」；、？！『』〈〉（）《》○【】]+'
df = pd.read_csv('./JudgeText/19612.csv',header=None,sep=',')
for i in range(len(df.columns)):
    text=df[i].tolist()
    for j in range(len(text)):
    #         print(type(text[j]))
        if type(text[j]) == float:
            continue
        
        strText += str(text[j])

zhText = re.findall(regex, strText)
cleanText = ''.join(zhText)
print(cleanText)
path = './totalText/19612.txt'
f = open(path, 'w', encoding='UTF-8')
print(cleanText, file=f)
f.close()
```

```python
# 合併所有判決書到一個txt檔案
dirs = os.listdir('.\JudgeText_RawData')
strText=""
for file in dirs:
    title = re.match(regex2,file)[0]
    df = pd.read_csv(f'./JudgeText_RawData/{file}',header=None,sep=',')
    for i in range(len(df.columns)):
        text=df[i].tolist()
        for j in range(len(text)):
    #         print(type(text[j]))
            if type(text[j]) == float:
                continue
            elif re.match(regex2,text[j]) == None:
                strText += str(text[j])
            else:
                continue

zhText = re.findall(regex, strText)
cleanText = ''.join(zhText)
path = './allText.txt'
f = open(path, 'w', encoding='UTF-8')
print(cleanText, file=f)
f.close()
```
